```python
# Q5
# part 1
marks={'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
print(marks)
# part 2
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Andy', 'Amy', 'James', 'Jules', 'Arthur'])
y=np.array([88, 66, 90, 55, 77])
plt.bar(x,y)
plt.xlabel('Students Names')
plt.ylabel('Students Grades')
plt.title('Names & Grades')
plt.show
```

```python
# part 3
import matplotlib.pyplot as plt
import numpy as np
z=np.array([88,66,90,55,77])
mylabels=['Andy', 'Amy', 'James', 'Jules', 'Arthur']
plt.pie(z, labels=mylabels)
plt.title('Names & Grades')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
# Q6
import matplotlib.pyplot as plt
import numpy as np
# multiple lines
x1=np.array([0,10])
y1=np.array([0,10])
x2=np.array([0,10])
y2=np.array([2,8])
x3=np.array([0,10])
y3=np.array([7,12])
plt.subplot(2,3,1)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.title('multiple lines')
# bar chart
x=np.array(['dog', 'cat', 'chi', 'bun', 'tur'])
y = np.array([2, 3, 30, 4, 1])
plt.subplot(2,3,2)
plt.bar(x,y, color='green')
plt.title('bar chart')
# pie chart
z=np.array([4,8,16,32,40])
plt.subplot(2,3,3)
plt.pie(z)
plt.title('pie chart')
# grid
x=np.array([2,3,4,5,6,7,19])
y=np.array([5,21,14,2,12,7,3])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.grid()
plt.title('grid')
# scatter plot
x=np.array([2,3,4,15,11,7,19])
y=np.array([5,21,14,2,12,7,3])
plt.subplot(2,3,5)
plt.scatter(x,y, color='hotpink')
plt.title('scatter plot')
#histogram plot
w=np.random.normal(100, 5, 200)
plt.subplot(2,3,6)
plt.hist(w, color='red')
plt.title('histogram plot')

plt.subplots_adjust(hspace=0.4)
plt.show()
```
